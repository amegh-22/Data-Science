```python
import numpy as np
import matplotlib.pyplot as plt

n_groups = 5
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)

fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, men_means, bar_width,alpha=opacity,color='g',label='Men')

rects2 = plt.bar(index + bar_width, women_means, bar_width,alpha=opacity,color='r',label='Women')

plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()
plt.show()


import numpy as np
import matplotlib.pyplot as plt

# Data
groups = ["G1", "G2", "G3", "G4", "G5"]
men = [22, 30, 33, 30, 26]
women = [25, 32, 30, 35, 29]

# Bar settings
x = np.arange(len(groups))
width = 0.35

# Plot
plt.bar(x, men, width, label="Men", color="green")
plt.bar(x + width, women, width, label="Women", color="red")

# Labels and title
plt.xlabel("Group")
plt.ylabel("Scores")
plt.title("Scores by Group")
plt.xticks(x + width / 2, groups)
plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

groups = ['G1', 'G2', 'G3', 'G4', 'G5']
men = [22, 30, 33, 30, 26]
women = [25, 32, 30, 35, 29]

x = np.arange(len(groups))
width = 0.4

plt.bar(x - width/2, men, width, label='Men', color='green')
plt.bar(x + width/2, women, width, label='Women', color='red')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by Group')
plt.xticks(x, groups)
plt.legend()
plt.show()
```

```python
#9
# import numpy as np
import matplotlib.pyplot as plt

label = []
values = []

# Read and parse the CSV
with open('medal.csv', 'r') as f:
    for line in f:
        parts = line.strip().split(',')
        if len(parts) == 2:
            label.append(parts[0])
            values.append(int(parts[1]))

# Plot the pie chart
plt.pie(values, labels=label, autopct='%1.1f%%')
plt.title("Medal Distribution")
plt.show()
```

```python
#10
# import numpy as np
import matplotlib.pyplot as plt

# Marks of 10 students
students = np.arange(1, 11)  # Student IDs from 1 to 10
math_marks = np.array([88, 92, 80, 89, 100, 80, 60, 100, 80, 34])
science_marks = np.array([35, 79, 79, 48, 100, 88, 32, 45, 20, 30])

# Create scatter plot
plt.scatter(students, math_marks, color='blue', label='Mathematics Marks')
plt.scatter(students, science_marks, color='orange', label='Science Marks')

# Add labels and title
plt.xlabel('Student Number')
plt.ylabel('Marks')
plt.title('Math vs Science Marks of 10 Students')

# Add legend
plt.legend(loc='upper right') 

# Show plot
plt.show()
```
