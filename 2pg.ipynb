```python
import numpy as np

def matrix(name):
    n = int(input(f"Enter the number of rows of matrix {name}: "))
    p = int(input(f"Enter the number of columns of matrix {name}: "))

    matri = []

    for i in range(n):
        row = []
        for j in range(p):
            value = int(input(f"Enter element at [{i}][{j}] of matrix {name}: "))
            row.append(value)
        matri.append(row)

    matrix_np = np.array(matri)
    return matrix_np



A = matrix("A")
print("Matrix A:")
print(A)
B = matrix("B")
print("Matrix B:")
print(B)

x=np.dot(A,B)
print("The Dot product")
print(x)

y=np.transpose(A)
z=np.transpose(B)
print("The transpose of first matrix")
print(y)
print("The transpose of second matrix")
print(z)

l=np.trace(A)
m=np.trace(B)
print("The trace of first matrix",l)
print("The trace of second matrix",m)

n=np.linalg.matrix_rank(A)
o=np.linalg.matrix_rank(B)
print("The rank of first matrix",n)
print("The rank of second matrix",o)

h=np.linalg.det(A)
i=np.linalg.det(B)
print("The determinant of first matrix")
print(h)
print("The determinant of second matrix")
print(i)

r=np.linalg.inv(A)
s=np.linalg.inv(B)
print("The inverse of first matrix")
print(r)
print("The inverse of second matrix")
print(s)
print("The inverse of the matrix:/n",np.linalg.inv(A))

t=np.linalg.eig(A)
u=np.linalg.eig(B)
print("The eigenvector of first matrix")
print(t)
print("The eigenvector of second matrix")
print(u)



```
